```python
import os
import pandas as pd

import sys
import os

# Añadir el directorio 'data_extraction' al sys.path
sys.path.append(os.path.abspath('../data_extraction'))

from CsvDataExtractor import CsvDataExtractor

# Crear una instancia de la clase CsvDataExtractor
csv_extractor = CsvDataExtractor(base_dir='../data')

# Obtener y mostrar la clasificación total desde el CSV
league_standings_total = csv_extractor.get_league_standings("total")
print("Clasificación Total:")
print(league_standings_total.head())

# Obtener y mostrar la clasificación como local desde el CSV
league_standings_home = csv_extractor.get_league_standings("home")
print("\nClasificación Local:")
print(league_standings_home.head())

# Obtener y mostrar la clasificación como visitante desde el CSV
league_standings_away = csv_extractor.get_league_standings("away")
print("\nClasificación Visitante:")
print(league_standings_away.head())

# Obtener y mostrar las estadísticas de todos los equipos desde el CSV
team_statistics = csv_extractor.get_all_teams_statistics()
print("\nEstadísticas de Equipos:")
print(team_statistics.head())

# Obtener y mostrar las estadísticas de todos los jugadores desde el CSV
all_player_stats = csv_extractor.get_all_player_statistics()
print("\nEstadísticas de Jugadores:")
print(all_player_stats.head())
```
